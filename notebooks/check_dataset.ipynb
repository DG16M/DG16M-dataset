```python
import numpy as np
import os 
import matplotlib.pyplot as plt
import trimesh
import seaborn as sea
import h5py 
import pandas as pd
from tqdm import tqdm

plt.style.use('ggplot')
```

```python
GRASP_PATH = '/scratch/dualarm/DG16M/dg16m/grasps/'
MESH_PATH = '/scratch/dualarm/DG16M/dg16m/meshes'
grasp_files = []

df = {
    'grasp_path': [],
    'grasp_file': [],
    'num_grasps': [],
    'passing_grasps': [],
    'failing_grasps': [],
}

failing_losses, passing_losses = [], []

for grasp_file in tqdm(os.listdir(GRASP_PATH)):
    if grasp_file.endswith('.h5'):
        grasp_files.append(os.path.join(GRASP_PATH, grasp_file))
        grasp_file = h5py.File(os.path.join(GRASP_PATH, grasp_file), 'r')
        grasps = grasp_file['grasps/grasps'][:]
        passing_indices = grasp_file['grasps/fc_passing_indices'][:]
        failing_indices = grasp_file['grasps/fc_failed_indices'][:]
        if len(passing_indices) == 0 or len(failing_indices) == 0:
            print(f"No passing grasps in {grasp_file}")
            continue

        loss_values = grasp_file['grasps/loss_values'][:]
        failing_losses.append(loss_values[failing_indices])
        passing_losses.append(loss_values[passing_indices]) 
        
        df['grasp_file'].append(grasp_files[-1].split('/')[-1])
        df['num_grasps'].append(len(grasps))
        df['passing_grasps'].append(len(passing_indices))
        df['failing_grasps'].append(len(failing_indices))
        df['grasp_path'].append(grasp_files[-1])
        
    else:
        print(f"{grasp_file} not supposed to use it!")
        
df = pd.DataFrame(df)
df
```
